## 2. 엑셀 읽기

### 1) 시트로 부터 셀의 값을 읽기

* 먼저 엑셀파일을 열어서 내용을 확인합시다.

이제 코드로 내용을 열어 보겠습니다.

* 셀 이름으로 열기

```python
type(wb)
```

```python
sheet1 = wb['Sheet1']

sheet1['A1'].value
```

```python
type(sheet1)
```

* 행, 열 번호로 읽기 (1부터 시작됩니다!)

```python
sheet1.cell(row = 1, column = 1).value
```

* 데이터가 있는 셀의 시작과 끝의 행을 메서드로 알아 낼 수 있습니다.

```python
print(sheet1.min_row, sheet1.max_row)
print(sheet1.min_column, sheet1.max_column)
```

* 실습①

```python
# Banana를 읽어 봅시다.
print(sheet1['A2'].value)
print(sheet1.cell(row=2, column=1).value)
```

```python
# green 을 읽어 봅시다.
sheet1.cell(3,2).value
```

```python
# for loop를 이용하여 과일 이름들을 하나씩 출력해 봅시다.
max_r=sheet1.max_row

for i in range(max_r):                  #행 : 0,1,2,3
    print(sheet1.cell(i+1,1).value)     #행 : 1,2,3,4
```

## 3. 엑셀에 쓰기

### 1) 셀에 값 수정하기
* 셀에 직접 값을 할당하면 수정이 된다.

```python
sheet1.cell(row = 1, column = 1).value = 'Apple'
```

```python
sheet1.cell(1,1).value='Apple'
```

```python
sheet1.cell(row = 1, column = 1).value
```

```python
sheet1.cell(1,1).value
```

* 그러나, 아직 파일에 쓴 것은 아니다.
* 아래와 같이 워크북 오브젝트를 저장해야 한다.


```python
wb.save('fruit2.xlsx')
```

- 같은 이름으로 저장 : 
    - 엑셀에서 저장하는 것이 아닌, 
    - 기존 파일에 똑같은 이름의 파일로 덮어 씌운다는 의미. 

* 실습 ②

```python
# 1행, 2열 셀의 값을 yellow 로 변경하시오.
sheet1.cell(1,2).value = 'yellow'

# 저장하고
wb.save('fruit2.xlsx')

# 잘 변경되었는지 불러와서 확인해 봅시다.

```

* 행/열 삭제하기
    * sheet.delete_rows(행번호) : 행 삭제
    * sheet.delete_cols(열번호) : 열 삭제
    
* 행/열 삽입
    * sheet.insert_rows(행번호) : 행 삽입
    * sheet.insert_cols(열번호) : 열 삽입

```python
wb = xl.load_workbook("animal.xlsx")
sheet1 = wb['Sheet1']

print(sheet1.cell(1,1).value, sheet1.cell(1,2).value)
print(sheet1.cell(2,1).value, sheet1.cell(2,2).value)
print(sheet1.cell(3,1).value, sheet1.cell(3,2).value)

# wb.save('animal.xlsx')
```

* 빈 행 추가하기

```python
sheet1.insert_rows(2) 

print(sheet1.cell(1,1).value, sheet1.cell(1,2).value)
print(sheet1.cell(2,1).value, sheet1.cell(2,2).value)
print(sheet1.cell(3,1).value, sheet1.cell(3,2).value)

# wb.save('animal.xlsx')
```

--- 빈 셀을 읽어오면 : None

* 행 삭제하기

```python
sheet1.delete_rows(2) 

print(sheet1.cell(1,1).value, sheet1.cell(1,2).value)
print(sheet1.cell(2,1).value, sheet1.cell(2,2).value)
print(sheet1.cell(3,1).value, sheet1.cell(3,2).value)

# wb.save('animal.xlsx')
```

변경사항은 wb.save() 를 실행해야만 적용됩니다.
