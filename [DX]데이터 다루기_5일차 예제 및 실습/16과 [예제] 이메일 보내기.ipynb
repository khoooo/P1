## 2. 이메일 보내기

### 1) 필요한 라이브러리 불러오기

```python
import smtplib
from email.mime.text import MIMEText
```

### 2) 메일 보내기 설정

```python
# 세션 생성
s = smtplib.SMTP('smtp.gmail.com', 587)
s.starttls()                                 # TLS 보안 시작

# 아래 부분을 여러분의 정보로 수정해야 합니다.
my_email='chlrlgh87@gmail.com'
app_pw='xvdtbyvqoplfuiir'
s.login(my_email, app_pw)        # 로그인 인증
```

### 3) 메일 내용 구성

```python
# ① 메일 내용 작성
msg = MIMEText('''메일 테스트 원 투 쓰리
![image.png](attachment:image.png)
''')

# ② 메일 제목 추가
msg['Subject'] = '파이썬으로 보내는 메일'
```

### 4) 메일 보내기

```python
s.sendmail(my_email, "lade7ten@naver.com", msg.as_string())
s.quit()                                    # 세션 종료
```
